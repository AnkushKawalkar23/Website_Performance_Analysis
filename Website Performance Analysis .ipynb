## Website Performace Analysis

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("D:\\Python\\data-export (1).csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
```

```python
df.head()
```

```python
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ['channel group', 'Datehour', 'Users', 'Sessions', 'Engaed Sessions', 'Average engagement time per session',	'Engaged sessions per user',	'Events per session',	'Engagement rate',	'Event count']
```

```python
df.head()
```

```python
# to check data type
df.info()
```

```python
# to chnage the data type and add column 
df['Datehour'] = pd.to_datetime(df['Datehour'], format= '%Y%m%d%H')
```

```python
df.info()
```

```python
df.head()
```

```python
# Convert all columns except 'channel group' and 'Datehour' to numeric
df.update(df.drop(columns=['channel group', 'Datehour']).apply(pd.to_numeric, errors='coerce'))

# Extract hour from 'Datehour' column
df['Hour'] = pd.to_datetime(df['Datehour']).dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Sessions and users overtime 

```python
sns.set(style='whitegrid')
```

```python
# Group by 'Datehour' and sum 'Sessions' and 'Users'
grouped = df.groupby('Datehour')[['Sessions', 'Users']].sum()

# Plot
plt.figure(figsize=(10, 5))
grouped.plot(ax=plt.gca())
plt.title("Sessions and Users over Time")
plt.xlabel("Datehour")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Total User by total user by Channel 

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='channel group', y='Users', estimator=np.sum)
plt.title('Total User by Channel')
plt.xticks(rotation =45)
plt.show()          
```

# Average Engagement Time By Channel 

```python
plt.figure(figsize=(8,5))
sns.barplot(data= df, x= 'channel group', y= 'Average engagement time per session', estimator=np.mean, palette='magma')
plt.title('Avg Engagement Time by Channel')
plt.xticks(rotation= 45)
plt.show()
```

```python
df.head()
```

# Engagement Rate Distribution by Channel 

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='channel group', y= 'Engagement rate', palette='magma')
plt.title('Engagement Rate Distribution by Channel')
plt.xticks(rotation= 45)
plt.show()
```

# Engaged Vs Non-Engaged Session 


```python
df.head()
```

```python
# 1. Group by and sum
session = df.groupby('channel group')[['Sessions', 'Engaed Sessions']].sum().reset_index()

# 2. Create Non-Engaged Sessions column
session['Non-Engaged Sessions'] = session['Sessions'] - session['Engaed Sessions']

# 3. Melt for plotting
session_melt = session.melt(
    id_vars='channel group',
    value_vars=['Engaed Sessions', 'Non-Engaged Sessions'],
    var_name='Session Type',
    value_name='Count'
)

# 4. Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=session_melt, x='channel group', y='Count', hue='Session Type')
plt.title('Engaged vs Non-Engaged Sessions by Channel Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Traffic by Hour and Channel 

```python
heatmap_df = df.groupby(['Hour','channel group'])['Sessions'].sum().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_df, cmap='YlGnBu', annot=True, fmt=".0f")
plt.title('Traffic by Hour and Channel Group')
plt.ylabel('Hour')
plt.xlabel('Channel Group')
plt.tight_layout()
plt.show()
```

# Corelation Between Engagement rate vs Sessions over time 

```python
df.head()
```

```python
# Group by Datehour and calculate average of both columns
df_correlation = df.groupby('Datehour')[['Engagement rate', 'Sessions']].mean().reset_index()

# Plot
plt.figure(figsize=(10, 5))

plt.plot(df_correlation['Datehour'], df_correlation['Engagement rate'], label='Engagement Rate', color='green')
plt.plot(df_correlation['Datehour'], df_correlation['Sessions'], label='Sessions', color='blue')

plt.title('Engagement Rate vs Sessions Over Time')
plt.xlabel('Datehour')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
